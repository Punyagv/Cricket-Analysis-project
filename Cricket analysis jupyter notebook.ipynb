```python
import pandas as pd
```

```python
df_match=pd.read_csv('dim_match_summary.csv')
```

```python
df_match.head()
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
match_ids_dict
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'mathch_id'},axis=1,inplace=True)
df_match.head()
```

```python
import pandas as pd

df_batting = pd.read_json('t20_wc_batting_summary.json')

all_records = []
for index, row in df_batting.iterrows():
    all_records.extend(row['battingSummary'])

```

```python
df_batting=pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting.tail()
```

```python
df_batting['out/not_out']=df_batting.dismissal.apply(lambda x:"out" if len(x)>0 else "not_out")
```

```python
df_batting.head(11)
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(10)
                
                
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x:x.replace('†',' '))
df_batting.head(20)
```

```python
match_ids_dict['Namibia Vs Sri Lanka']
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
```

```python
df_batting.head()
```

```python
df_batting.shape
```

```python
df_batting.to_csv('fact_batting_summary.csv',index=False)
```
